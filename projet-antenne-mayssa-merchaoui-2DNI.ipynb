```python
# afficher le 30 premiére ligne de data frame
dfmay.head(30)
```

```python

X=dfmay[['Numéro de mesure','Niveau_global']].to_numpy()
print(X)
y=dfmay['target'].to_numpy()
print(y)
```

```python
#pour l'apprentissage ona diviser x en 'X_train', 'X_test' et y en 'y_train', 'y_test'
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#scaler.fit(X_train)
#from sklearn.neighbors import KNeighborsClassifier
#logreg1 = KNeighborsClassifier(n_neighbors = 2)
#classifier.fit(X_train, y_train)
from sklearn.linear_model import LogisticRegression 
#Create a l.Reg Classifier
logreg1 = LogisticRegression()
#Train the model using the training sets
logreg1.fit(X_train, y_train)
y_pred=logreg1.predict(X_test)
```

#### **** je faire un  teste avnec 3 algorthime  pour celle de deux algorithme de prediction (algorithme de KNN et algorithme  )il on donne des resultat inferieur a celle de regression logistique  

```python
#afficher  la matrice de confusion
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
